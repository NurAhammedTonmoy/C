```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Load the dataset
df = pd.read_csv('your_dataset.csv')  # replace with your actual CSV file path

# Make a copy
df_clean = df.copy()

# --- Step 1: Handle structured strings ---
# Parse display_resolution like "1080 x 2400"
df_clean[['res_width', 'res_height']] = df_clean['display_resolution'].str.extract(r'(\d+)\s*x\s*(\d+)').astype(float)

# Parse dimensions like "160.8 x 78.1 x 7.65 mm"
dims = df_clean['dimensions'].str.extract(r'(\d+\.?\d*)\s*x\s*(\d+\.?\d*)\s*x\s*(\d+\.?\d*)')
df_clean[['dim_length', 'dim_width', 'dim_thickness']] = dims.astype(float)

# Parse video_resolution like "4K@30fps" or "1080p@60fps"
df_clean['video_fps'] = df_clean['video'].str.extract(r'@(\d+)').astype(float)

# Convert launch_date to year
df_clean['launch_date'] = pd.to_datetime(df_clean['launch_date'], errors='coerce')
df_clean['launch_year'] = df_clean['launch_date'].dt.year.fillna(0).astype(int)

# --- Step 2: Drop unneeded or now-parsed columns ---
df_clean.drop(columns=[
    'display_resolution', 'dimensions', 'video', 'launch_date', 'year'
], inplace=True)

# --- Step 3: Clean battery_size ---
# Extract numeric part like "5000 mAh" → 5000
df_clean['battery_size'] = df_clean['battery_size'].str.extract(r'(\d+)').astype(float)

# --- Step 4: Handle object (categorical) columns ---
label_enc = LabelEncoder()

# Fill NA for categorical features
df_clean.fillna("Unknown", inplace=True)

for col in df_clean.select_dtypes(include='object').columns:
    df_clean[col] = label_enc.fit_transform(df_clean[col].astype(str))

# --- Step 5: Convert remaining columns to numeric ---
df_clean = df_clean.apply(pd.to_numeric, errors='ignore')

# --- Final check ---
print("Final shape:", df_clean.shape)
print(df_clean.head())

# Optionally, save it to a new CSV
df_clean.to_csv("cleaned_phone_data.csv", index=False)
```
